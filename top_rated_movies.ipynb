```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Top_Rated_Movies.csv', parse_dates=['release_date'])
df.head()
```

```python
#create a column only for years
df['release_date'] = pd.to_datetime(df['release_date'])
df['year'] = df['release_date'].dt.year
```

```python
#which year were the most top rated movies release
top_years = df['year'].value_counts().head()
plt.pie(top_years, labels=top_years.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('top 5 years', fontsize=14)
plt.show()
```

```python
print('2017 has the most movies on the list')
```

```python
#are there any duplicates
df.duplicated().sum()
```

```python
#drop the duplicates
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.lineplot(data=df, x='year', y= 'vote_average')
plt.grid(True)
plt.title("Year vs vote average")
plt.show()
```

```python
sns.regplot(data=df, x='year', y= 'vote_average')
```

```python
#what is the average vote
df['vote_average'].mean()
```

```python
average_vote = df['vote_average'].mean()

sns.lineplot(data=df, x='year', y='vote_average')

# Add a horizontal line for the average vote
plt.axhline(y=average_vote, color='red', linestyle='--', label='Average Vote')

plt.legend()

plt.grid(True)
plt.show()
```
