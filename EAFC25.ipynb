```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load the data
df = pd.read_csv('male_players.csv', index_col = 0 )
df.head()
```

```python
#get to know the data better
#see all the columns we have
df.columns
```

```python
#get the stat of the data
df.describe()
```

```python
#the number of rows and columns we have
df.shape
```

```python
#check if there is any missing data
df.isnull().sum()
```

```python
#which team is the best in EAFC25
#betwen Real madrid and Manchester City
man_city = df[df['Team'] == 'Manchester City']
print('Manchester City overall average is', man_city['OVR'].mean())
real_madrid = df[df['Team'] == 'Real Madrid']
print('Real Madrid overall average is', real_madrid['OVR'].mean())
print('Real Madrid have better players according to EAFC25, sounds a bit baised')
```

```python
# who is the strongest player in the game
# who is the strongest player in the game
strongest_player = df[df['Strength'] == df['Strength'].max()][['Name', 'Rank']]
strongest_player
strongest_player
```

```python
#since there are many strong players lets put there physicality to account
strongest_player = df[df['Strength'] == df['Strength'].max()][['Name', 'Rank','Strength','PHY']]
strongest_player.sort_values(by = 'PHY', ascending = False)
```

```python
#who is the most skillful player with 
#good finishing, speed, ball control and weak foot

best = df[df['Skill moves'] == df['Skill moves'].max()][['Name', 'OVR', 'Skill moves', 'Finishing','Weak foot']]
best.sort_values(by = ['OVR', 'Skill moves', 'Finishing', 'Weak foot'], ascending=False).head()
```

```python
#who is the best english player
#best brazilian player
best_eng_player = df.loc[(df['Nation'] == 'England')]
jude = best_eng_player.head(1)
best_bra_player = df.loc[(df['Nation'] == 'Brazil')]
best_bra_player.head(1)
new_df = pd.concat([best_bra_player.head(), best_eng_player.head()])
new_df
```

```python
sns.barplot(data=new_df, x ='SHO', y = 'Name', palette = 'Blues', estimator = np.median, ci= 0, hue = 'SHO')
plt.show()
```
