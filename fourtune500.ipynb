```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('Fortune_500.csv')
data
```

```python
#check for missing values
data.isnull().sum()
```

```python
print('There are no missing values')
```

```python
#check for duplicates
dup = data.duplicated().any()
print('Are there any duplicates: ', dup)
```

```python
#check how many sectors there are in the dataset
data['Sector'].value_counts()
```

```python
#get the sectors with the lowest revenue
sns.countplot(data=data,x='Sector')
plt.xticks(rotation=45)
plt.show()
```

```python
data.columns
```

```python
#Which states has the most companies
sns.countplot(data['Hqstate'])
```

```python
#which person is the ceo of multiple companies
ceo_counts = data['Ceo'].value_counts()

ceo_multiple_companies = ceo_counts[ceo_counts > 1]
print(ceo_multiple_companies)
```

```python
top_5_companies = data.nlargest(5, 'Totshequity')

plt.figure(figsize=(8, 8))
plt.pie(top_5_companies['Totshequity'], labels=top_5_companies['Title'], autopct='%1.1f%%')
plt.title('Top 5 Companies by Total Equity')
plt.show()
```

```python
data.columns
```

```python
#which industry makes the most profits
industry_profits = data.groupby('Industry')['Profits'].sum()
most_profitable_industry = industry_profits.idxmax()
max_profit = industry_profits.max()

print(f"The most profitable industry is {most_profitable_industry} with a total profit of {max_profit}")
```

```python
plt.figure(figsize=(12, 6))
industry_profits.plot(kind='line', marker='o', color='green')
plt.title('Total Profits by Industry')
plt.xlabel('Industry')
plt.ylabel('Total Profits')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
