```python
# which team had the most losses
losses = data.loc[data['losses'].idxmax(), 'team']
print("the team with the most loses is" ,losses)
```

```python
#which team had the least goals conceded
goal_con = data.loc[data['goals_conceded'].idxmin(),'team']
print("the team with least goals conceded", goal_con)
```
